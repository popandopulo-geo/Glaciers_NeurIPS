```python

pathOrigin = os.path.join(path, "datasets", name, "rawData","Temperatures")
#print(pathOrigin)
os.chdir(pathOrigin)

files = glob.glob(os.path.join(pathOrigin, '*'))
data = []
for i in range(len(files)):
    data.append(alignment.openData(files[i]))
    break
print("data loaded")

data = data[0]
print('Printing the data for 2013')
print('type: ',type(data))
print('len: ', len(data))
```
